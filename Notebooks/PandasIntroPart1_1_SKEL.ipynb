## Here is a function that I found to decrease the memory of your data set by picking a smaller data type (e.g., 8 bit float from 64 bit float)
## It isn't much of an issue for our in-class data, but handy for larger data sets.

```python
# Function to decrease data usage in notebook (from Baek Kyun Shin)
def downcast(df, verbose=True):
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        dtype_name = df[col].dtype.name
        if dtype_name == 'object':
            pass
        elif dtype_name == 'bool':
            df[col] = df[col].astype('int8')
        elif dtype_name.startswith('int') or (df[col].round() == df[col]).all():
            df[col] = pd.to_numeric(df[col], downcast='integer')
        else:
            df[col] = pd.to_numeric(df[col], downcast='float')
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose:
        print('{:.1f}% Compressed'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
```

```python
df_movies = downcast(df_movies)
```

```python
df_movies.info()
```

## Reduced from 1.1 MB (1100 KB) to 0.768 MB (768 KB) a 28.6% reduction in memory! Mostly from moving from float64 and int64 to float32 and int32.
